```python
# 노드와 엣지 데이터 경로
node_csv = r'C:\Users\1102w\Desktop\대학강의자료\프로젝트\쿠다 심화프로젝트\marker_id.csv'
edge_csv = r'C:\Users\1102w\Desktop\대학강의자료\프로젝트\쿠다 심화프로젝트\엣지_최종_리얼.csv'

# 노드 데이터 읽기 (CSV)
df = pd.read_csv(node_csv, encoding='cp949')

# 열 이름 정리
df.columns = df.columns.str.strip()  # 공백 제거

# 불필요한 열 제거 (예: Unnamed 열)
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# 'id' 열이 없으면 생성
if 'id' not in df.columns:
    df['id'] = range(1, len(df) + 1)

# 엣지 데이터 읽기 (CSV)
df3 = pd.read_csv(edge_csv, encoding='utf-8-sig')

# 지도 중심을 첫 번째 노드의 좌표로 설정
map_center = [df['위도'].iloc[0], df['경도'].iloc[0]]  # 첫 번째 행의 위도, 경도 사용
m = folium.Map(location=map_center, zoom_start=15)

# 각 노드에 ID를 popup으로 표시하는 마커 추가
for idx, row in df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],  # 노드 위치
        popup=f"ID: {row['id']}",  # 팝업에 ID 표시
        icon=folium.Icon(color='blue', icon='info-sign')  # 마커 색상 변경
    ).add_to(m)

# 엣지를 지도에 추가
for _, row in df3.iterrows():
    # 시작 노드와 끝 노드의 좌표 가져오기
    id_start = row['id_start']
    id_end = row['id_end']
    
    # 시작 노드와 끝 노드의 위치 찾기
    start_node = df[df['id'] == id_start]
    end_node = df[df['id'] == id_end]
    
    # 두 노드의 좌표 추출
    if not start_node.empty and not end_node.empty:
        start_coords = [start_node['위도'].values[0], start_node['경도'].values[0]]
        end_coords = [end_node['위도'].values[0], end_node['경도'].values[0]]
        
        # 선 색상 설정: crosswalk == 1이면 빨간색, 아니면 파란색
        line_color = 'red' if row['crosswalk'] == 1 else 'blue'
        
        # 지도에 선 추가 (폴리라인)
        folium.PolyLine(
            locations=[start_coords, end_coords],
            color=line_color,  # 선 색상
            weight=2,  # 선 두께
            opacity=0.7  # 선 투명도
        ).add_to(m)

# 지도 저장 및 출력
m.save("nodes_and_edges_with_crosswalk.html")  # HTML로 저장
m
```
